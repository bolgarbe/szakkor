## Wave Function Collapse (WFC) algoritmus

```julia
using Plots, Images, FileIO, StatsBase
```

```julia
img = load("valami.bmp")
```

```julia
s = 3
tile_counts = countmap([img[i:i+s-1,j:j+s-1] for i in 1:size(img,1)-s, j in 1:size(img,2)-s])
tiles,w,n   = collect(keys(tile_counts)),collect(values(tile_counts)),length(tile_counts)

tiles
```

```julia
rcomp = [all(t1[1:s,s] .== t2[1:s,1]) for t1 in tiles, t2 in tiles]
lcomp = [all(t1[1:s,1] .== t2[1:s,s]) for t1 in tiles, t2 in tiles]
dcomp = [all(t1[s,1:s] .== t2[1,1:s]) for t1 in tiles, t2 in tiles]
ucomp = [all(t1[1,1:s] .== t2[s,1:s]) for t1 in tiles, t2 in tiles];
```

```julia
function collapse!(c,i,j)
    cw   = c[:,i,j] .* w
    tile = wsample(1:n,cw)
    c[:,i,j] .= false
    c[tile,i,j] = true
end;
```

```julia
valid = [(i,j) -> j<N, (i,j) -> j>1, (i,j) -> i<N, (i,j) -> i>1]
comp  = [rcomp,lcomp,dcomp,ucomp]
dir   = [(0,1),(0,-1),(1,0),(-1,0)]

function propagate!(c,i,j)
    stack = [(i,j)]
    while length(stack)>0
        i,j = pop!(stack)

        for d in 1:4
            if valid[d](i,j)
                i_,j_ = (i,j) .+ dir[d]
                comp_ = dropdims(any(c[:,i_,j_] .* comp[d][c[:,i,j],:]',dims=2),dims=2)
                if any(c[:,i_,j_] .!= comp_)
                    c[:,i_,j_] .= comp_
                    push!(stack,(i_,j_))
                end
            end
        end
    end
end;
```

```julia
N,M = 24,24
c   = ones(Bool,n,N,M)
collapsed = dropdims(sum(c,dims=1),dims=1) .== 1

anim = Animation()
ctr  = 0
while !all(collapsed)
    w2  = c.*w ./ sum(c.*w,dims=1)
    ent = -dropdims(sum(c.*(w2.*log.(w2)),dims=1),dims=1) .+ randn(N,M)/100 .+ collapsed .* 1000
    m   = argmin(ent)
    collapse!(c,m[1],m[2])
    propagate!(c,m[1],m[2])
    
    collapsed .= dropdims(sum(c,dims=1),dims=1) .== 1
    
    ctr += 1
    if ctr % 5 == 0
        out = zeros(typeof(img[1]),(N*s,M*s))
        for i in 1:N, j in 1:M
            out[(i-1)*s+1:i*s,(j-1)*s+1:j*s] .= mean(tiles[findall(c[:,i,j])])
        end
        plot(out)
        frame(anim)
    end
end
gif(anim)
```
